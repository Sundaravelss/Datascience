```python
iris.dropna(how='all',inplace=True)
iris
```

```python
iris_x=iris.iloc[:,:-1]
iris_x
```

```python
from sklearn.preprocessing import StandardScaler

irs_std=StandardScaler().fit_transform(iris_x)

irs_std
```

```python
#covariance

mean_vector=np.mean(irs_std,axis=0)

covariance=((irs_std-mean_vector).T).dot(irs_std-mean_vector)

covariance=covariance/(irs_std.shape[0]-1)

covariance
```

```python
#correlation matrix

correlation=np.corrcoef(irs_std.T)
correlation
```

```python
eigvalues,eigvec=np.linalg.eig(correlation)

print(eigvalues,eigvec)
```

```python
#checking

if np.array_equal(correlation,(eigvec.dot(np.diag(eigvalues)).dot(np.linalg.inv(eigvec)))):
    print('success')
else:
    print('fail')
```

```python
eigvec.dot(eigvalues*np.identity(4)).dot(np.linalg.inv(eigvec))
```

```python
#plot

tot=sum(eigvalues)
print(tot)

explained_variance=[(i/tot)*100 for i in sorted(eigvalues,reverse=True)]
print(explained_variance)
```

```python
li=[5,3,2]

per=[i/sum(li)*100 for i in li]
per
```

```python
cum_explained_variance=np.cumsum(explained_variance)
cum_explained_variance
```
